```python
import os
import json
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline

file = open (os.path.relpath('../test/WithSolution.json', os.path.abspath('')))
data = json.load(file)

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
num = np.array([e["n"] for e in data])
mean = np.array([np.mean(e["t"]) for e in data])
std = np.array([np.std(e["t"], ddof=1) for e in data])

plt.plot(num, mean, color='green', marker='o',linewidth=3, markersize=10, label="Média")
plt.plot(num, std, color='blue', marker='o',linewidth=3, markersize=10, label="Desvio padrão")

plt.ylabel('Tempo de execução (microssegundos)',fontsize=14)
plt.xlabel('Tamanho N do "lado" do Sudoku',fontsize=14)

plt.legend()
plt.show()
```
